```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('classic')
%matplotlib inline

# Increase default figure and font sizes for easier viewing.
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['font.size'] = 14
```

```python
# Read in the Boston housing data.
housing_csv = 'boston_housing_data.csv'
housing = pd.read_csv(housing_csv)
```

```python
housing[['ZN', 'INDUS']].plot(figsize=(12,8), style={'ZN': '-g', 'INDUS': '--b'});
```

```python
df = pd.DataFrame(np.random.randn(10, 4), 
                  columns=['col1', 'col2', 'col3', 'col4'],
                  index=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'])
```

```python
ax = df.plot(kind='bar', figsize=(15,3));

# Set the title.
ax.set_title('Plot Thingy', fontsize=21,y=1);

# Move the legend.
ax.legend(loc=3);

# x-axis labels
ax.set_ylabel('y-axis info', fontsize=16);

# y-axis labels
ax.set_xlabel('x-axis info', fontsize=16);
```

```python
ax = df.plot(kind='barh', figsize=(15,3));

# Set the title.
ax.set_title('Plot Thingy', fontsize=21,y=1);

# Move the legend.
ax.legend(loc=3);

# x-axis labels
ax.set_ylabel('y-axis info', fontsize=16);

# y-axis labels
ax.set_xlabel('x-axis info', fontsize=16);
```

```python
ax = df.plot(kind='barh', figsize=(15,3));

# Set the title.
ax.set_title('Plot Thingy', fontsize=21,y=1);

# Move the legend.
ax.legend(loc=1);

# x-axis labels
ax.set_ylabel('y-axis info', fontsize=16);

# y-axis labels
ax.set_xlabel('x-axis info', fontsize=16);
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(housing['CRIM'], housing['TAX'], c='blue', alpha=0.7)

plt.xlabel('CRIM')
plt.ylabel('TAX')
plt.title('Scatter Plot of Variables with Positive Correlation')

plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(housing['ZN'], housing['NOX'], c='red', alpha=0.7)

plt.xlabel('ZN')
plt.ylabel('NOX')
plt.title('Scatter Plot of Variables with Negative Correlation')

plt.show()
```
